```python
import seaborn as sns
df2 = sns.load_dataset("tips")
```

```python
df2.head(2)
```

```python
df2.iplot(kind="scatter", x="total_bill", y='tip', mode='markers', size=8)
```

```python
df2.iplot(kind='bar', x='total_bill', y='tip', colors='indigo')
```

```python
df2.iplot(kind='box')
```

```python
df3 = pd.DataFrame(df2["total_bill"],df2['tip'], df2['size'])
df3.iplot(kind='surface', colorscale="rdylbu")
```
